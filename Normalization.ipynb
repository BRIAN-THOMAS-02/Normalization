# Normalization

***

***

When features of the input Dataset have large differences between their ranges or simply when they are measure in different measurement units

The Main Intention of Normalization is that it converts the : <br>
<b> 1) Values to a range between 0 & 1 </b> <br>
<b> 2) Keep the Range of the Original Values just as is </b> <br>

Basically we are scaling down the values of our Dataset to a scale which will be pretty much similar to all the feature values, but most impotantly keeping the <b> RANGE SAME AS IN DATASET </b>

***

Using this Formula we replace all the values individually we have in input with the Transformed/Recalculated values of the Normalized Values

<b> 1) First we calculate the Min of the Feature/Input Column </b> <br>
    
<b> 2) Then we Calculate the Max of the Feature/Input Column </b> <br>
    
<b> 3) Finally we Substitute the Min & Max in the Normalization Formula for each individual value and calculate the Normalized Values Individually for Each Value. </b> <br> <br>


NOTE : 

When we <b> Subtract the Smallest Value from the Min, <i>(the Numerator becomes 0) </i> </b>. Hence we get <b>(0) as Output. </b> <br>
When we <b> Subtract the Largest Value from the Max, <i> (the Numerator == Denomiator )</i> </b>. Hence we get <b> (1) as Output. </b> <br>

<b> Hence, all our Normalized Values will ALWAYS be in the range 0 - 1, </b> <br>

***

# We'll check out the Basic Implementation

```python
dataset_0 = [10,5,6,1,3,7,9,4,8,2]
dataset_1 = [1,99,789,5,6859,541,94142,7,50826,35464]
n = len(dataset_1)

min_dataset_1 = min(dataset_1)
max_dataset_1 = max(dataset_1)

print("Min(dataset_1) : ", min_dataset_1 )
print("Max(dataset_1) : ", max_dataset_1 )
```

***

## Finding the Normalized Values for Each Input Value in the Dataset

```python
final_normalization = []

print("\nNormalization = (Current_Value - Min) / (Max - Min)\n")

for i in dataset_1:
    normalization = (i - min_dataset_1) / (max_dataset_1 - min_dataset_1 )
    final_normalization.append("{:.20f}".format(normalization))
    
    print("For", i)
    print("\t (", i,"-",min_dataset_1,") / (", max_dataset_1, "-", min_dataset_1,")")
    print("\tNormalization(", i, ") : {:.20f}".format(normalization))
    print("\n")
```

## Comparing the Input Values and the Normalized Values in the Dataset

```python
print("  Original DataSet   |         Normalization ")
print()
for i in range (len(dataset_1)):
    print("         ", dataset_1[i], "         |     ", final_normalization[i])
```

```python
plt.scatter(dataset_0, dataset_1, label= "stars", color= "blue", marker= "*", s=40)
    
plt.xlabel('Index')
plt.ylabel('Original Values')

plt.title('Graph of Original Values')
plt.legend()

plt.show()
```

```python
plt.scatter(dataset_0, final_normalization, label= "stars", color= "blue", marker= "*", s=30)
  
plt.xlabel('Index')
plt.ylabel('Normalization Values')

plt.title('Graph of Normalized Values')
plt.legend()
  
plt.show()
```
